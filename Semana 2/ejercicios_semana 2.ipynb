## Ejercicio 1: Basic Complex Number Operations
1. Manually calculate and then verify using Python the result of `(3 + 4j) + (1 - 2j)`, `(5 + 6j) * (7 - 8j)`, and `(2 - 3j) / (1 + 4j)`.
2. Find the conjugate and modulus of `(3 - 4j)`.

```python
# Definición de números complejos y operaciones básicas

# Suma de números complejos
defined_complex1 = complex(3, 4)  # Número complejo 3 + 4j
defined_complex2 = complex(1, -2)  # Número complejo 1 - 2j
print(f"El resultado de la suma es: {defined_complex1 + defined_complex2}")

# Multiplicación de números complejos
multiplicand_complex = complex(5, 6)  # Número complejo 5 + 6j
multiplier_complex = complex(7, -8)  # Número complejo 7 - 8j
print(f"Resultado de la multiplicación: {multiplicand_complex * multiplier_complex}")

# División de números complejos
# (2 - 3j) / (1 + 4j)
numerator_complex = complex(2, -3)  # Número complejo 2 - 3j
denominator_complex = complex(1, 4)  # Número complejo 1 + 4j
print(f"Resultado de la división: {numerator_complex / denominator_complex}")

# Cálculo del módulo de un número complejo
# |3 - 4j|
modulus_complex = complex(3, -4)  # Número complejo 3 - 4j
print(f"Resultado del módulo: {abs(modulus_complex)}")
```

## Ejercicio 2: Visualizing Complex Operations
1. Write a Python function to plot a complex number on the complex plane. Use it to plot `(3 + 4j)` and its conjugate.
2. Extend the function to show addition and multiplication of two complex numbers graphically.

```python
import cmath
import matplotlib.pyplot as plt
import numpy as np

# Función para graficar un número complejo y su conjugado
def plot_complex_numbers(complex_num, conjugate_num):
    ax = plt.subplot()
    
    # Graficar el número complejo y su conjugado
    ax.plot([0, complex_num.real], [0, complex_num.imag], label="Número Complejo")
    ax.plot([0, conjugate_num.real], [0, conjugate_num.imag], label="Conjugado")
    
    # Configurar límites de la gráfica
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    
    # Etiquetar los ejes
    ax.set_xlabel("Real")
    ax.set_ylabel("Imaginary")
    
    # Configurar cuadrícula y ejes
    ax.axhline(0, color="black", linewidth=0.5)
    ax.axvline(0, color="black", linewidth=0.5)
    ax.grid(True)
    
    # Título y ajustes
    plt.title("Vector y su Conjugado")
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    complex_num = complex(3, 4)  # Definiendo el número complejo 3 + 4j
    conjugate_num = complex_num.conjugate()  # Calculando su conjugado
    
    plot_complex_numbers(complex_num, conjugate_num)

```

## Exercise 4: Creating a Julia Set
1. Implement a Julia set generator. Use a constant like `-0.4 + 0.6j` for the iteration `z = z*z + constant`.
2. Explore how changing the constant changes the pattern of the Julia set.

```python
import matplotlib.pyplot as plt

# Definición de parámetros para el conjunto de Julia
julia_constant = complex(-0.4, 0.6)  # Constante utilizada en la iteración
max_distance = 1 + abs(julia_constant)  # Rango máximo de valores complejos
max_iterations = 100  # Número máximo de iteraciones

# Definición del rango de coordenadas
x_start, y_start = -1, 1  # Esquina superior izquierda
x_end, y_end = 1, -1  # Esquina inferior derecha
resolution = 500  # Resolución de la cuadrícula

# Cálculo de la cantidad de puntos en cada eje
num_rows = abs(int((y_end - y_start) * resolution))
num_cols = abs(int((x_end - x_start) * resolution))

# Generación de coordenadas en el espacio complejo
coord_x = [x_start + ((x_end - x_start) / num_cols) * i for i in range(num_cols)]
coord_y = [y_start + ((y_end - y_start) / num_rows) * i for i in range(num_rows)]

# Inicialización de la matriz de valores del conjunto de Julia
julia_set = [[0 for _ in coord_x] for _ in coord_y]

# Cálculo del conjunto de Julia
for y in range(len(coord_y)):
    for x in range(len(coord_x)):
        complex_num = complex(coord_x[x], coord_y[y])
        iterations = 0
        
        while abs(complex_num) < max_distance and iterations < max_iterations:
            iterations += 1
            complex_num = complex_num**2 + julia_constant
        
        julia_set[y][x] = 0 if iterations == max_iterations else iterations

# Configuración y visualización del gráfico
fig, ax = plt.subplots()
ax.set_aspect("equal")
plot = ax.pcolormesh(coord_x, coord_y, julia_set, cmap="magma", shading="auto")
plt.colorbar(plot)
plt.title("Conjunto de Julia")
plt.show()
```
